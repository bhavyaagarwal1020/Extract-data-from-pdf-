```python
!pip install pypdf
from pypdf import PdfReader
```

```python
reader = PdfReader('ProjectsampleReport4th.pdf')

#counting number of pages
print(len(reader.pages))

#accessing 1st page
page = reader.pages[0]

#extract the text from grabbed page
#print(page.extract_text())

#goto every page and extract text
for i in range(len(reader.pages)):
  page = reader.pages[i]
  print(page.extract_text())


for i in range(len(reader.pages)):
  page = reader.pages[i]
  for j in page.images:
    with open(j.name, 'wb') as f:
      f.write(j.data)
```
